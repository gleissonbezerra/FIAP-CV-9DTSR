# 2. Exemplo

```python
import cv2
import dlib
import numpy as np
from google.colab.patches import cv2_imshow

# Carregar detector de faces do Dlib
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")  # Baixe esse modelo

# Carregar imagem e converter para escala de cinza
img = cv2.imread("pessoas.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detectar faces
faces = detector(gray)

for face in faces:
    # Obter os 68 pontos da face detectada
    landmarks = predictor(gray, face)

    #TAREFA: Tratar os landmarks e plotar nas faces demonstrando a detecção

cv2_imshow(img)
```
