Este notebook demonstra como calcular o mapa de disparidade (depth map) usando um par estéreo e o algoritmo StereoBM do OpenCV.

Fonte das imagens: dataset *Tsukuba* (Middlebury Stereo Vision)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
import os
```

# 2. Exemplo com Imagens de Laboratório

```python
left_path = "tsukuba-imL.png"
right_path = "tsukuba-imR.png"

# Carregar imagens em escala de cinza
imgL = cv2.imread(left_path, cv2.IMREAD_GRAYSCALE)
imgR = cv2.imread(right_path, cv2.IMREAD_GRAYSCALE)

# Criar objeto StereoBM e computar disparidade
stereo = cv2.StereoBM_create(numDisparities=64, blockSize=15)
disparity = stereo.compute(imgL, imgR)

# Exibir resultados
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.title("Imagem Esquerda")
plt.imshow(imgL, cmap='gray')
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Imagem Direita")
plt.imshow(imgR, cmap='gray')
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Mapa de Disparidade")
plt.imshow(disparity, cmap='inferno')
plt.axis("off")
plt.tight_layout()
plt.show()
```
